```python
import pandas as pd
import numpy as np
import random

# Extract address data ensuring all required columns are present
clients_addresses = clients_df[['Street', 'State']].copy()
clients_addresses['City'] = np.nan  # Add a City column filled with NaNs

office_expenses_addresses = office_expenses_df[['Address', 'State', 'City']].rename(columns={'Address': 'Street'})

home_listings_addresses = home_listings_df[['Street', 'City']].copy()
home_listings_addresses['State'] = np.nan  # Add a State column filled with NaNs

# Combine address data ensuring no rows are dropped due to missing information
combined_addresses = pd.concat([clients_addresses, office_expenses_addresses, home_listings_addresses], ignore_index=True)

# Remove duplicates
combined_addresses = combined_addresses.drop_duplicates()

# Function to separate street and unit based on the first comma or the first three words
def separate_street_and_unit(street):
    if ',' in street:
        parts = street.split(',', 1)
        return parts[0].strip(), parts[1].strip()
    parts = street.split()
    if len(parts) > 3:
        return ' '.join(parts[:3]), ' '.join(parts[3:])
    return street, ''

# Apply the function to extract and update Street and Unit
combined_addresses[['Street', 'Unit']] = combined_addresses['Street'].apply(lambda x: pd.Series(separate_street_and_unit(x)))

# Standardize and fill in missing values
combined_addresses['Country'] = 'USA'
combined_addresses['Zip_Code'] = combined_addresses['Zip_Code'].astype(str) if 'Zip_Code' in combined_addresses.columns else ''

# Preserve existing Zip Codes if available and generate random 5 digits for missing ones
combined_addresses['Zip_Code'] = combined_addresses['Zip_Code'].replace('', np.nan)
# Change here: Generate the random zip codes as a pandas Series instead of a numpy array
zip_codes_to_fill = pd.Series(np.random.randint(10000, 99999, size=len(combined_addresses))).astype(str)
combined_addresses['Zip_Code'].fillna(zip_codes_to_fill, inplace=True)

# Generating values for address_id
combined_addresses.insert(0, 'address_id', range(1, 1 + len(combined_addresses)))

print(combined_addresses)
```

```python
# Merge office data with address data using a left join
office_data = office_expenses_addresses.copy()
office_data['Street'] = office_data['Street'].apply(lambda x: x.split(',')[0].strip())

# Assume initial phone numbers are not available and set to 'N/A'
office_data['phone_number'] = 'N/A'

import re

# Function to standardize street addresses
def standardize_address(address):
    address = address.lower().strip()  # Normalize case and strip whitespace
    address = re.sub(r'\bstreet\b', 'st', address)
    address = re.sub(r'\broad\b', 'rd', address)
    address = re.sub(r'\bavenue\b', 'ave', address)
    address = re.sub(r'\bsuite .*', '', address)  # Remove suite numbers
    return address

# Apply standardization to both dataframes
office_data['Street'] = office_data['Street'].apply(standardize_address)
combined_addresses['Street'] = combined_addresses['Street'].apply(standardize_address)

# Merge data with improved matching
joined_data = office_data.merge(
    combined_addresses[['Street', 'address_id']],
    on='Street',  # Ensure this key matches the right column
    how='left'  # Use 'left' join to keep all records from office_data
)

# Check for NaN in 'address_id' and log details
nan_address_ids = joined_data[joined_data['address_id'].isna()]

# Assuming joined_data is your final merged DataFrame from previous steps
unmatched_addresses = joined_data[joined_data['address_id'].isna()]

# Log details of unmatched addresses for review or processing
if not unmatched_addresses.empty:
    print("There are unmatched addresses that need to be reviewed or added to the addresses database.")
    print(unmatched_addresses[['Street', 'State', 'City']])
```

```python
# Pass the SQL statements that create all tables
stmt = """
CREATE TABLE addresses (
    address_id SERIAL PRIMARY KEY,
    street VARCHAR(255),
    unit VARCHAR(50),
    city VARCHAR(100),
    state VARCHAR(50),
    zip_code CHAR(5) NOT NULL,
    country VARCHAR(50)NOT NULL
);

CREATE TABLE offices (
    officeid SERIAL PRIMARY KEY,
    addressid INT,
    officename VARCHAR(255),
    phonenumber VARCHAR(20),
    FOREIGN KEY (addressid) REFERENCES addresses(address_id)
);

CREATE TABLE employees (
    employeeid SERIAL PRIMARY KEY,
    officeid INT,
    firstname VARCHAR(100) NOT NULL,
    lastname VARCHAR(100) NOT NULL,
    role VARCHAR(100),
    email VARCHAR(255) NOT NULL,
    phonenumber VARCHAR(20),
    salary DECIMAL(10, 2) NOT NULL,
    FOREIGN KEY (officeid) REFERENCES offices(officeid)
);

CREATE TABLE clients (
    clientid SERIAL PRIMARY KEY,
    firstname VARCHAR(100) NOT NULL,
    lastname VARCHAR(100) NOT NULL,
    email VARCHAR(255) NOT NULL,
    phonenumber VARCHAR(250),
    clienttype VARCHAR(50) NOT NULL
);

CREATE TABLE amenities (
    amenityid SERIAL PRIMARY KEY,
    amenitytype VARCHAR(60) NOT NULL UNIQUE
    );

CREATE TABLE neighborhoodfeatures (
    featureid SERIAL PRIMARY KEY,
    featuretype VARCHAR(150),
    featurename TEXT
    );

CREATE TABLE home_listings (
    homeid BIGINT PRIMARY KEY,
    addressid INT NOT NULL,
    sellerid INT NOT NULL,
    listingdate DATE NOT NULL,
    price DECIMAL(12, 2) NOT NULL,
    status VARCHAR(50) NOT NULL,
    bedroomcount INT NOT NULL,
    bathroomcount INT NOT NULL,
    squarefootage INT NOT NULL,
    FOREIGN KEY (addressid) REFERENCES addresses(address_id),
    FOREIGN KEY (sellerid) REFERENCES clients(clientid)
);

CREATE TABLE home_id_amenities (
    id SERIAL PRIMARY KEY,
    homeid BIGINT NOT NULL,
    amenityid INT NOT NULL,
    FOREIGN KEY (homeid) REFERENCES home_listings(homeid),
    FOREIGN KEY (amenityid) REFERENCES amenities(amenityid)
);

CREATE TABLE home_neighborhood_features (
    id SERIAL PRIMARY KEY,
    homeid BIGINT NOT NULL,
    featureid INT NOT NULL,
    FOREIGN KEY (homeid) REFERENCES home_listings(homeid),
    FOREIGN KEY (featureid) REFERENCES neighborhoodfeatures(featureid)
);

CREATE TABLE past_transactions (
    transactionid SERIAL PRIMARY KEY,
    homeid BIGINT NOT NULL,
    buyerid INT NOT NULL,
    sellerid INT NOT NULL,
    employeeid INT NOT NULL,
    date TIMESTAMP NOT NULL,
    amount DECIMAL(12, 2) NOT NULL,
    transactiontype VARCHAR(50) NOT NULL,
    FOREIGN KEY (homeid) REFERENCES home_listings(homeid),
    FOREIGN KEY (buyerid) REFERENCES clients(clientid),
    FOREIGN KEY (sellerid) REFERENCES clients(clientid),
    FOREIGN KEY (employeeid) REFERENCES employees(employeeid)
);

CREATE TABLE financial_records (
    recordid SERIAL PRIMARY KEY,
    date TIMESTAMP NOT NULL,
    type VARCHAR(100) NOT NULL,
    notes TEXT NULL,
    transactionid INT NOT NULL,
    employeeid INT NOT NULL,
    officeid INT NOT NULL,
    FOREIGN KEY (transactionid) REFERENCES past_transactions(transactionid),
    FOREIGN KEY (employeeid) REFERENCES employees(employeeid),
    FOREIGN KEY (officeid) REFERENCES offices(officeid)
);

CREATE TABLE client_inquiries (
    inquiryid SERIAL PRIMARY KEY,
    clientid INT NOT NULL,
    date TIMESTAMP NOT NULL,
    inquirytype VARCHAR(100),
    description TEXT,
    FOREIGN KEY (clientid) REFERENCES clients(clientid)
);

CREATE TABLE appointments (
    appointmentid INT PRIMARY KEY,
    clientid INT NOT NULL,
    employeeid INT NOT NULL,
    date DATE NOT NULL,
    time TIME NOT NULL,
    purpose VARCHAR(255) NOT NULL,
    notes TEXT,
    FOREIGN KEY (clientid) REFERENCES clients(clientid),
    FOREIGN KEY (employeeid) REFERENCES employees(employeeid)
);

CREATE TABLE user_favorites (
    userid INT,
    homeid INT,
    favoritedondate DATE NOT NULL,
    PRIMARY KEY(userid, homeid),
    FOREIGN KEY (userid) REFERENCES clients(clientid),
    FOREIGN KEY (homeid) REFERENCES home_listings(homeid)
);

CREATE TABLE open_houses (
    openhouseid SERIAL PRIMARY KEY,
    homeid INT NOT NULL,
    date DATE NOT NULL,
    starttime TIME,
    endtime TIME,
    notes TEXT,
    FOREIGN KEY (homeID) REFERENCES home_listings(homeID)
);

CREATE TABLE property_maintenance_requests (
    maintenancerequestid SERIAL PRIMARY KEY,
    homeid BIGINT NOT NULL,
    requestdate TIMESTAMP NOT NULL,
    issuedescription TEXT NOT NULL,
    urgencylevel VARCHAR(50) NOT NULL,
    resolutionstatus VARCHAR(100) NOT NULL,
    FOREIGN KEY (homeid) REFERENCES home_listings(homeid)
);


"""

# Execute the statement to create tables
connection.execute(stmt)
```
